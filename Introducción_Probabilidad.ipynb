La **teoría de la probabilidad** es la parte de las matemáticas que se encarga del estudio de los fenómenos o experimentos aleatorios. Por experimento aleatorio entenderemos todo aquel experimento que cuando se le repite bajo las mismas condiciones iniciales, el resultado que se obtiene no siempre es el mismo.

La **teoría de la probabilidad** es la base de la inferencia estadística y un instrumento esencial en el análisis de la variabilidad.

El término **probabilidad** proviene de lo probable, o sea, de aquello que es más posible que ocurra, y se entiende como el mayor o menor grado de posibilidad de que un evento aleatorio ocurra, expresado en una cifra entre 1 (posibilidad total) y 0 (imposibilidad absoluta), o bien en porcentajes entre el 100% o el 0%, respectivamente.

##Espacio muestral o de resultados
El conjunto de todos los resultados posibles de un experimento estadístico se llama **espacio muestral** (EM) y se representa con el símbolo $\Omega$.

Cada resultado en un espacio muestral se llama **elemento** o miembro del espacio muestral, o simplemente punto muestral. Si el espacio muestral tiene un número finito de elementos, podemos listar los miembros separados por comas y encerrados en paréntesis.

Por ejemplo, para el experimento de lanzar una moneda, el EM se puede escribir como:
$$\Omega = \{C,S\}$$

###Complemento de un evento
El complemento de un evento $A$ con respecto al espacio muestral $\Omega$ es el subconjunto de todos los elementos que no están en $A$. Denotamos el complemento de $A$ mediante $A’$ (o $A^c$).

#### Ejemplo complemento
Sea $R$ el evento de que una carta roja se seleccione de una baraja ordinaria de 52 cartas, y sea el EM toda la baraja. Entonces $R’$ es el evento de que la carta seleccionada de la baraja no sea roja sino negra.

###Intersección de dos eventos
La intersección de dos eventos $A$ y $B$, denotada mediante el símbolo $A \cap B$, es el evento que contiene a todos los elementos que son comunes a $A$ y a $B$.

####Ejemplo intersección
Sea $P$ el evento de que una persona seleccionada al azar, mientras cena en un restaurante de moda, se conecta a facebook, y sea $Q$ el evento de que la persona tenga más de 45 años de edad. Entonces el evento $P \cap Q$ es el conjunto de todas las personas con facebook en el restaurante que tienen más de 45 años de edad.

###Unión de dos eventos
La unión de dos eventos $A$ y $B$, que se denota mediante el símbolo $A \cup B$ , es el evento que contiene todos los elementos que pertenecen a $A$ o a $B$ o a ambos.

####Ejemplo de unión de dos eventos
Sea $C$ el evento de que un empleado seleccionado al azar de una compañía de seguros fume cigarrillos. Sea $B$ el evento de que el empleado seleccionado ingiera bebidas alcohólicas. Entonces el evento $C \cup B$ es el conjunto de todos los empleados que beben alcohol o fuman, o que hacen ambas cosas.

### Eventos excluyentes
Dos eventos $A$ y $B$ son mutuamente excluyentes o disjuntos si la intersección de ellos es vacía, es decir, si $A$ y $B$ no tienen elementos en común.

#### Ejemplo eventos excluyentes

## Prinicipios fundamentales del conteo
En lo que respecta a técnicas de conteo, tenemos dos principios importantes:
* El principio de adición
* El principio de multiplicación

###El principio de adición (o)
Si un evento o suceso $A$ ocurre de $n$ maneras y otro $B$ ocurre de $m$ maneras, luego:

Número de maneras en que puede ocurrir el evento $A$ o el evento $B$ es: $n + m$.

**Nota:** Un evento o suceso ocurre de una forma o de otra, más no de ambas formas a la vez (no sucede en simultaneo).

###El principio de multiplicación (y)
Si un evento $A$ ocurre de $n$ maneras diferentes seguido de otro evento $B$ que ocurre de $m$ maneras distintas, entonces:

Número de maneras en que puede ocurrir el evento $A$ y el evento $B$ es: $n \times m$.

**Nota:** Los eventos ocurren uno a continuación de otro, originando un evento compuesto.

##Definición clásica de probabilidad
Sea $\Omega$ un espacio muestral finito que contiene $N$ eventos simples, y sea $A$ un evento que puede darse de $n$ maneras distintas; es decir, que al realizar un experimento hay $N$ resultados posibles de los cuales $n$ son favorables al evento $A$. La probabilidad de que ocurra el evento $A$ está dada por:

$$P(A) = \frac{\textit{resultados favorables}}{\textit{resultados posibles}} = \frac{n}{N}$$

Si relacionamos la definición precedente con el repaso de la Teoría de Conjuntos, podemos afirmar que la probabilidad de que se dé el evento $A$ está dada por el cociente entre la cantidad de elementos del conjunto favorables al evento A y el número de elementos del espacio muestral $\Omega$, siendo estos últimos igualmente probables.

###Ejemplo de probabilidad
En un juego se lanzan dos dados de manera simultanea. ¿Cúal es la probabilidad de que la suma de los dos dados sea igual a 7? 

```
import itertools
import numpy as np
listas = [range(1,7),range(1,7)]
jugadas = np.array(list(itertools.product(*listas))) # producto cartesiano
```

[producto cartesiano](https://es.wikipedia.org/wiki/Producto_cartesiano) es la operación entre dos conjuntos cuyo resultado son todos los pares ordenados que pueden formarse. El primer elemento del par pertence al primer conjunto mientras que el segundo par corresponde al segundo conjunto.

¿Cuál es el evento A?

```
jugadas7 = jugadas[np.sum(jugadas, axis=1)==7]
jugadas7
```

```
p7 = len(jugadas7)/len(jugadas)
print(f'La probabilidad del evento A es: {p7}')
```

###Definición frecuentista
La Definición Frecuentista de probabilidad surge debido a la existencia de fenómenos aleatorios
en los cuales no se puede determinar con precisión la probabilidad clásica de cada evento simple,
es decir, que no podemos precisar cuántos resultados favorables a un evento existen y/o cuántos
resultados posibles hay.

Ejemplos de esto son los siguientes:
* Un jefe de control de calidad desea determinar la probabilidad
de que un artículo sea defectuoso.
* Un fanático está interesado en la probabilidad de que su
equipo de fútbol gane.
* un profesor que quiere saber la probabilidad de que sus alumnos
aprueben.

Para estimar la probabilidad de cada uno de esos eventos, se recurre a la segunda manera de
definir a la probabilidad, utilizando la frecuencia relativa de ocurrencia de los mismos.

Sea $K$ el número de veces que se observa un fenómeno determinado, y sea
$k$ el número de veces en que
ocurre un resultado favorable al evento $A$. La probabilidad de ocurrencia del evento $A$
es la frecuencia relativa observada cuando el número total de observaciones crece indefinidamente:
$$P(A) = \lim_{k \to \infty}\frac{k}{K}$$
Es decir, que la probabilidad de que ocurra el evento A resulta de dividir el número de veces que A, efectivamente, apareció entre el número de veces que se intentó el experimento.

####Ejemplo
Un surtido de dulces contiene seis mentas, cuatro chicles y tres chocolates. Si una persona hace una selección aleatoria de uno de estos dulces, calcular la probabilidad de sacar a) una menta y b) un chicle o un chocolate.

Si $M$, $T$ y $C$ representan los eventos de que la persona seleccione, respectivamente, una menta, un chicle y un chocolate. El número total de dulces es 13, los cuales tiene la misma probabilidad de ser seleccionados.

a.) Como seis de los 13 dulces son mentas,$P(M)= 6/13$

b.) Como siete de los 13 dulces son chicles o chocolates, $P(T \cup C) = 7/13$ 

###Definición subjetiva
La Definición Subjetiva de probabilidad está relacionada con el grado de creencia que tiene quien
lleva a cabo un experimento respecto de la probabilidad de ocurrencia del mismo.

En este caso la probabilidad de un evento depende del observador, es decir, según lo que el observador conoce del fenómeno en estudio. Puede parecer un tanto informal y poco serio esta forma de definir la probabilidad de un evento, sin embargo en muchas situaciones es necesario recurrir a un experto para tener por lo menos una idea vaga de cómo se comporta el fenómeno de nuestro interés y saber si la
probabilidad de un evento es alta o baja. Por ejemplo, ¿cuál es la probabilidad de que nuestro equipo favorito de futbol gane en su próximo partido? Ciertas
circunstancias internas del equipo, las condiciones del equipo rival o cualquier otra condición externa, son elementos que sólo algunas personas conocen y que podrían darnos una idea más exacta de esta probabilidad.

Esta probabilidad suele llamarse también probabilidad a *priori*, ya que refleja el grado de creencia antes de que se realice cualquier prueba empírica.

Este tipo de análisis de problemas es lo que se conoce como **Análisis Bayesiano**, mediante el
cual se modifican las probabilidades subjetivas (a priori) utilizando el Teorema de Bayes. La tarea consiste en analizar la información suministrada por los
resultados de algún tipo de experimento, para obtener probabilidades condicionadas a dicha información.

##Axiomas de probabilidad
Dentro de la teoría de la probabilidad ilustran tres postulados básicos de la probabilidad, a los que se conoce como axiomas de probabilidad, lo que en lenguaje matemático significa que son proposiciones que por su carácter evidente no requieren demostración. Constituyen, por decirlo de alguna manera, “las reglas del juego”, sin importar si estamos trabajando una probabilidad subjetiva o empírica, o si seguimos los postulados de la probabilidad clásica.

Estos axiomas, que constituyen el cimiento de la teoría moderna de probabilidades, fueron propuestos por el matemático ruso [A. N. Kolmogorov](https://es.wikipedia.org/wiki/Andr%C3%A9i_Kolmog%C3%B3rov) y se expresan de manera formal en los siguientes términos.
1. Para todo evento $A$, $P(A) \geq 0$ (no hay probabilidades negativas)
2. Para el espacio muestral $\Omega$, $P(\Omega) = 1$. (ningún evento tiene una probabilidad mayor a 1)
3. Dados dos eventos, $A$ y $B$, $P(A \cup B) = P(A) + P(B) - P(A \cap B)$


image.png

A partir de estos axiomas, se tienen otros resultados importantes, tales como:
* $P(\phi) = 0$, donde $\phi$ representa el conjunto vacio
* $P(A^c) = 1 - P(A)$ o lo que es lo mismo $P(A) + P(A^c) = 1$
